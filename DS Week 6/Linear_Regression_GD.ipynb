```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data
x=np.array([1,2,3,4,5,7])
y=np.array([7.1,8.9,10.7,13.2,15.8,19.4])
```

```python
y.shape[0]
```

```python
#Mean Square Error
def loss(y, y_pred):
    L = (y_pred-y)**2
    L = np.sum(L)/y.shape[0]
    return L
```

```python
def theta_update(theta, x, y, y_pred):
    temp=(y_pred-y)*x
    temp=alpha*np.sum(temp)/y.shape[0]
    return (theta-temp)
```

```python
#initialization
theta_0=4
theta_1=1
alpha=0.1
Loss=[]
for i in range(25):
    y_pred=theta_0+theta_1*x
    print(y_pred)
    Loss.append(loss(y,y_pred))
    theta_0=theta_update(theta_0,1,y,y_pred)
    theta_1=theta_update(theta_1,x,y,y_pred)
```

```python
plt.plot(Loss, '-ro')
plt.xlabel("iteration")
plt.ylabel("Loss")
```

```python
theta_0, theta_1
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, theta_0 + theta_1 * x, color='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.show()
```
