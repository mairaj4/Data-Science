# Pandas
- The name "Pandas" is derived from the term "Panel Data."
- Panel data refers to multidimensional structured data, typically involving observations over time.
- Wes McKinney, the creator of Pandas, chose this name to reflect the core functionality of the library, which is to provide data structures and tools for working with structured data, particularly time series and tabular data sets.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Pandas provides two types of classes for handling data:

- Series: a one-dimensional labeled array holding data of any type such as integers, strings, Python objects etc.

```python
s = pd.Series([1, 3, 5, np.nan, 6, 8])
s.shape, s
```

- DataFrame: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

```python
df2 = pd.DataFrame({ "A": 1.0, "B": pd.Timestamp("20130102T120505"), "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"), "E": pd.Categorical(["test", "train", "test", "train"]),"F": "foo",})
df2.head()
```

```python
dates = pd.date_range("20130101", periods=4)
df2.index=dates
df2.index
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['Target']=iris.target
df.head()
```

# locating particular records in a dataframe using indexing and slicing

```python
df.iloc[50:100]
```

```python
df.index = np.arange(1,300,2)
df.loc[50:100]
```

```python
df.iloc[50:100]
```

```python
df.reset_index(inplace=True, drop=True)
df.head()
```

# Feature selection using correlation

```python
import seaborn as sns
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True)
```

# In detail analysis of data based on groups of data

```python
temp=df.groupby(['Target'])
temp.boxplot(figsize=(12,12), rot=30);
```

# Handling Missing Values

```python
df.iloc[:11:2,1]=np.nan
df.head()
```

```python
df.fillna(df.iloc[::,1].mean())
```

# Encoding - convert categorical to number format

```python
df2 = pd.DataFrame({'Roll No':[1,2,3,4,5,6,7,8], 'Department':['CS','DS','CIS','DS','EL','AI', 'DS', 'CS']})
df2.head()
```

```python
df2['Department'].map({'CS':0,'DS':1, 'CIS':2, 'AI':3, 'EL':4}) # simple encoding
```

```python
pd.get_dummies(df2) # one-hot encoding
```
